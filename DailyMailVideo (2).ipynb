### A 2-3 paragraph write up reporting your results and answering your research question

Through the analysis we conducted of this TikTok video, discouragement and critique of Taylor Swift suing Donald Trump appeared more substantial and frequent than explicit support for a lawsuit. The number of comments coded as supporting a lawsuit was notably lower, as people often made fun of Swift for being dramatic in the comments. 

However, many commenters defended Swift, framed Trump’s use of her music as distasteful, or expressed approval of the idea that she should take legal action. However, not many people expressed specific support for Trump. Pro-Trump comments were often more critical of Swift, rather than directly supportive of Trump. Many TikTok users also reacted with jokes, confusion, or off-topic comments.
